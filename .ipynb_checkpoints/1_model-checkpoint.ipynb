# cosine similarity

```python
movies=pd.read_csv("./datasets/combined.csv")
```

# overall

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,MinMaxScaler,OneHotEncoder,OrdinalEncoder

cf=ColumnTransformer(transformers=[
    ('ORencoding',OrdinalEncoder(),['hit/flop','release_day','release_month']),
    ('OHencoding',OneHotEncoder(),['original_language','status']),
    ('std',StandardScaler(),['budget','revenue','popularity','vote_average','vote_count',
                             'runtime','profit_per','box-off'])
],remainder='passthrough')
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

pc=PCA(n_components=3)
ppl=Pipeline([
    ('cf',cf)
])
trans=ppl.fit_transform(df)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(trans)
def Recommend(movie):
    index=movies[movies['original_title']==movie].index
    curr=similarity[index]
    top5=np.argsort(curr)[0,:][::-1][1:6]
    return movies[movies.index.isin(top5)]['original_title']
Recommend('Avatar')
```

# Scale

```python
df=pd.read_csv('./datasets/scale.csv')
```

```python
scale=df.drop(['id','genres','keywords','spoken_languages','original_title','overview',
              'status','release_date','release_day','production_companies','production_countries','cast','crew',
              'release_month','tagline','release_year','runtime','original_language'],axis=1)
```

```python
scale.head()
```

```python
scale['budget']=scale['budget']/1000000
scale['revenue']=scale['revenue']/1000000
scale['box-off']=scale['box-off']/1000000
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,MinMaxScaler,OneHotEncoder,OrdinalEncoder

cf=ColumnTransformer(transformers=[
    ('ORencoding',OrdinalEncoder(),['hit/flop']),
    ('std',StandardScaler(),['budget','popularity','revenue','vote_average','vote_average','box-off','profit_per'])
],remainder='passthrough')
```

```python
from sklearn.pipeline import Pipeline

ppl=Pipeline([
    ('std',cf)
])
X=ppl.fit_transform(scale)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(X)
def Recommend(movie):
    index=movies[movies['original_title']==movie].index
    curr=similarity[index]
    top5=np.argsort(curr)[0,:][::-1][1:6]
    return movies[movies.index.isin(top5)]['original_title']
Recommend('Avatar')
```

```python
import pickle 

with open('./models/scale.pkl','wb') as f:
    pickle.dump(similarity,f)
```
