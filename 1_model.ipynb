```python
movies = pd.read_csv("./datasets/combined.csv")
```

# overall

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import (
    StandardScaler,
    MinMaxScaler,
    OneHotEncoder,
    OrdinalEncoder,
)

cf = ColumnTransformer(
    transformers=[
        ("ORencoding", OrdinalEncoder(), ["hit/flop", "release_day", "release_month"]),
        ("OHencoding", OneHotEncoder(), ["original_language", "status"]),
        (
            "std",
            StandardScaler(),
            [
                "budget",
                "revenue",
                "popularity",
                "vote_average",
                "vote_count",
                "runtime",
                "profit_per",
                "box-off",
            ],
        ),
    ],
    remainder="passthrough",
)
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

pc = PCA(n_components=3)
ppl = Pipeline([("cf", cf)])
trans = ppl.fit_transform(df)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(trans)


def Recommend(movie):
    index = movies[movies["original_title"] == movie].index
    curr = similarity[index]
    top5 = np.argsort(curr)[0, :][::-1][1:6]
    return movies[movies.index.isin(top5)]["original_title"]


Recommend("Avatar")
```
